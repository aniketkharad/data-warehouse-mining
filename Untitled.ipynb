```python
import pandas as pd
df = pd.read_csv('Bank_details.csv', index_col=0)
```

```python
df.drop('Surname', axis=1, inplace=True)
df.set_index("CustomerId", inplace=True)
df
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# df['Geography'] = df['Geography'].astype('category')
# df['Geography_values'] = df['Geography'].cat.codes
# df
# can use either this or the one below
# results in the same transformation
```

```python
df['Geography'] = df['Geography'].astype('category')
df['Gender'] = df['Gender'].astype('category')

labelencoder = LabelEncoder()
df['Gender_skl'] = labelencoder.fit_transform(df['Gender'])
df['Geography_skl'] = labelencoder.fit_transform(df['Geography'])
df
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
# enc_df = pd.DataFrame(enc.fit_transform(df[ ['Geography_skl']]).toarray())
# df = df.join(enc_df)

X = df[ ['Geography_skl']]
enc.fit(X)
y = enc.transform(X).toarray()
enc_df = pd.DataFrame(y)
enc_df
```

```python
df = df.join(enc_df)
df
```
